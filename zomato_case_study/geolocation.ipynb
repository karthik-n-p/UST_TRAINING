#### Fetching location details using geocode

```python
from geopy.geocoders import Nominatim
import pandas as pd
locations = pd.DataFrame({"Name":zomato_df['location'].unique()})
locations['Name'] = locations['Name'].apply(lambda x : "Bangalore " + str(x))
geolocator = Nominatim(user_agent="app")

lat_long = []
for location in locations['Name']:
    location = geolocator.geocode(location ,timeout=10)
    if location is None:
        lat_long.append(np.nan)
    else:
        geo = (location.latitude, location.longitude)
        lat_long.append(geo)
```

```python
locations['geo_loc'] = lat_long
locations.head()
```

```python
from geopy.geocoders import Nominatim
import pandas as pd
import numpy as np


filtered_df = zomato_df[zomato_df['cuisines'].str.contains("North Indian", case=False, na=False)]


locations = pd.DataFrame({"Name": filtered_df['location'].unique()})
locations['Name'] = locations['Name'].apply(lambda x: "Bangalore " + str(x))

# Initialize the geolocator
geolocator = Nominatim(user_agent="app")

# List to store latitude and longitude
lat_long = []

# Geocode each location
for location_name in locations['Name']:
    try:
        location = geolocator.geocode(location_name, timeout=10)
        if location is None:
            lat_long.append(np.nan)
        else:
            geo = (location.latitude, location.longitude)
            lat_long.append(geo)
    except Exception as e:
        print(f"Error geocoding {location_name}: {e}")
        lat_long.append(np.nan)


```

```python
s = s.dropna(subset=['geo_loc'])

# Extract latitude and longitude from the 'geo_loc' column
def extract_lat_long(geo_str):
    # Handle case where geo_str is not a string
    if isinstance(geo_str, str):
        # Remove parentheses and split by comma
        geo_str = geo_str.strip('()')
        lat, lon = geo_str.split(', ')
        return float(lat), float(lon)
    else:
        return np.nan, np.nan

# Apply the extraction function and create new columns
s[['lat', 'long']] = s['geo_loc'].apply(extract_lat_long).apply(pd.Series)

# Print results
print(s)
```
