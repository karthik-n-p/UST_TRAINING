#### 7. Handling missing values

```python
import numpy as np
#replace '?' by 'nan' and update dataframe
raw_data.replace('?',np.nan,inplace=True)
```

```python
raw_data.isnull().sum() / raw_data.shape[0] * 100# returns the total number of missing values from  each column name.
```

```python
#handling missing values
```

```python
raw_data.drop(['normalized-losses','symboling'],axis =1,inplace=True)
```

```python
raw_data['price'].unique()
```

```python
raw_data['price'].dtype
```

```python
raw_data['price'].replace('?',np.nan ,inplace=True)
```

```python
vehicle_df = raw_data.copy()
```

```python
vehicle_df['price'].median()
```

```python
raw_data['price'].fillna(vehicle_df['price'].median())
```

#### 13. Merge

```python
df_series1= pd.Series({
    'us':300000,
    'Germany':20000,
    'Japan' : 50000,},
    name='Population')

df_series2= pd.Series({
    'us':300,
    'Germany':200,
    'Japan' : 500,},
    name='gdp')
    
```

```python
df1=pd.DataFrame(df_series1).reset_index()
df1
```

```python
df1.columns = ['Country','gdp_per_capita'] # rename existing columns
df1
```

```python
df3=pd.DataFrame(df_series2).reset_index()
df3
```

```python
df3.columns = ['Country','Population']
df3
```

```python
pd.merge(df1,df3, on='Country',how='inner')
```
