#### Data type and plots
        Different types of plots are used to visualize data depending on the nature of the data and the insights you want to gain. Here’s a rundown of various plots and the types of data they’re suited for:

### 1. **Histogram**
   - **Purpose:** Shows the distribution of a single continuous variable.
   - **Data Suited For:** Numerical data, especially to visualize frequency distributions and understand the shape of the data distribution (e.g., normal, skewed).

### 2. **Bar Chart**
   - **Purpose:** Compares quantities across different categories.
   - **Data Suited For:** Categorical data, where each bar represents a category and its height represents the quantity or frequency (e.g., sales by product category).

### 3. **Box Plot (Box-and-Whisker Plot)**
   - **Purpose:** Displays the distribution of data based on quartiles and highlights outliers.
   - **Data Suited For:** Continuous data, to show the median, quartiles, and potential outliers (e.g., test scores across different classes).

### 4. **Scatter Plot**
   - **Purpose:** Shows the relationship between two continuous variables.
   - **Data Suited For:** Paired data points, to identify correlations or patterns between two variables (e.g., height vs. weight).

### 5. **Line Graph**
   - **Purpose:** Displays trends over time or ordered categories.
   - **Data Suited For:** Data points that follow a sequential order, such as time series data (e.g., stock prices over time).

### 6. **Pie Chart**
   - **Purpose:** Shows proportions of a whole.
   - **Data Suited For:** Categorical data, where each slice represents a category’s contribution to the total (e.g., market share of different companies).

### 7. **Heatmap**
   - **Purpose:** Visualizes data density or magnitude with color gradients.
   - **Data Suited For:** Data matrices or correlation matrices, to identify patterns or areas of concentration (e.g., correlation between different variables).

### 8. **Violin Plot**
   - **Purpose:** Combines a box plot with a density plot to show distribution and probability density.
   - **Data Suited For:** Continuous data, particularly to compare distributions across different categories (e.g., comparing the distribution of exam scores across different classes).

### 9. **Pair Plot**
   - **Purpose:** Visualizes pairwise relationships in a dataset.
   - **Data Suited For:** Multivariate data, to show scatter plots for all pairs of variables and histograms for each variable (e.g., relationships among several features in a dataset).

### 10. **Area Chart**
   - **Purpose:** Displays cumulative totals over time or categories, emphasizing volume.
   - **Data Suited For:** Time series or ordered data where you want to highlight the magnitude of change over time (e.g., cumulative sales over months).

Choosing the right plot depends on what you want to analyze or communicate about your data. Each type provides different insights and helps in understanding the data from various perspectives.
